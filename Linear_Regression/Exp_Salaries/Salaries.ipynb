```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
sal = pd.read_csv("Exp_Salaries.csv")
sal.head()
```

```python
lr = LinearRegression()
x = sal[['Years_Previous_Experience']]
y = sal['Salary']
lr.fit(x,y)
#print(lr.intercept_)
#print(lr.coef_)
y_pred = lr.predict(x)
print(r2_score(y,y_pred))
```
