```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from pca import pca
import seaborn as sns
```

```python
milk = pd.read_csv("C:\\Aalesh and Mandar\\Datasets\\milk.csv", index_col = 0)
```

```python
scalar = StandardScaler()
```

```python
scaled_milk = scalar.fit_transform(milk)
```

```python
pca = PCA()
```

```python
prin_comp = pca.fit_transform(scaled_milk)
```

```python
print(milk.shape)
```

```python
print(prin_comp.shape)
```

```python
pd_PC = pd.DataFrame(prin_comp,columns = ["pc1","pc2","pc3","pc4","pc5"])
```

```python
pd_PC.var()
```

```python
print(pca.explained_variance_)
```

```python
print(pca.explained_variance_ratio_ * 100)
```

```python
ys=np.cumsum(pca.explained_variance_ratio_ * 100)
xs=np.arange(1,6)
plt.plot(xs, ys)
plt.title("scree plot")
plt.xlabel("Principle Componants")
plt.ylabel("Cumulative percentage variation explained")
```
