```python
import ipywidgets as widgets
from IPython.display import display
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
import numpy as np
import os
os.chdir(r"C:\Users\admin\Desktop\Mandar")
df = pd.read_csv("Insure_auto.csv")
```

```python
df.head()
```

```python
y = df["Operating_Cost"]
X = df[['Home']]
regressor = LinearRegression()
regressor.fit(X, y)

print(regressor.coef_)
print(regressor.intercept_)
```

```python
def predict_on_test(x1):
    test_data = np.array([[x1]])
    prediction = regressor.predict(test_data)
    y1_str = "Operating_Cost Sales = " + str(prediction[0])
    display(y1_str);
```

```python
min_lstat = df['Home'].min()
max_lstat = df['Home'].max()
print((min_lstat,max_lstat))
```

```python
w = widgets.interactive(predict_on_test,  
                        x1 = widgets.FloatSlider(value=10.0,min=min_lstat,
                                                 max=max_lstat,step=0.05,
                                                 description='Operating_Cost:',disabled=False,
                                                 continuous_update=False))
display(w)
```
